```python
data = pd.read_csv("C:\\Users\\kisku\\Desktop\\Data_csv.csv")
```

```python
likes = data["likes"]
ratio = data["ratio"]
view_count = data["view_count"]
```

```python
plt.title("Trending Youtube Videos")
plt.xlabel("view_count")
plt.ylabel("likes")
plt.xscale("log")
plt.yscale("log")

plt.scatter(view_count,likes, s=100, edgecolor="black",  c=ratio, cmap='summer', linewidth=1,alpha=0.75 )

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')
```
